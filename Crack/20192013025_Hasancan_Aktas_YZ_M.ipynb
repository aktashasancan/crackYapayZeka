```python
#Veri Setine eklenen Kütüphaneler

from scipy import stats
import imageio as imo #Resimler için kullanılan kütüphane
import pandas as np #Veri işleme ve analizi yapar
import numpy as np #listelerden daha hızlı olduğu için kullanılan kütüphane
import os # klasör ve dosyaları okumak için lazım olan kütüphane
import matplotlib.pyplot as plt #Veri görselleştirmek için kullanılan kütüphane
import cv2 #Tüm görüntü işleme kullanılan görüntüleme kütüphanesi
```

```python
#Veri setinin dosya yolunu ve sınıflarını veriyoruz.

DATADIR="C://Users//Hasan//Desktop//DataSet"
CATEGORİES=["Negative","Positive"]
Negative=["Negative"]
Positive=["Positive"]
```

```python
#Veri setinde kaç tane resim olduğunu gösterir

training_data = []
def create_training_data():
    for category in CATEGORİES:
        path = os.path.join(DATADIR,category)
        class_num = CATEGORİES.index(category)
        for img in os.listdir(path):
            try:
                img_array = cv2.imread(os.path.join(path,img))
                training_data.append([img_array,class_num])
            except Exception as e:
                pass
create_training_data()
print(len(training_data))
```

```python
#Her bir resmin yüksekliğini ve genişliğini veren dizi setidir.

img_height =[]
img_width = []
for category in CATEGORİES:
    path = os.path.join(DATADIR , category)
    for img in os.listdir(path):
        img_array = cv2.imread(os.path.join(path,img))
        img_array = cv2.cvtColor(img_array, cv2.COLOR_BGR2RGB)
        img_height.append(img_array.shape[0])
        img_width.append(img_array.shape[1])
```

```python
##TÜM GÖRSELLERİN YÜKSEKLİK DAĞILIM HİSTOGRAMI
plt.hist(img_height ,color='#0c457d' , bins = 10, histtype = "bar" , align = "mid")
plt.xlabel("Height")
plt.ylabel("Resim Adedi")
plt.title("Görsellerin Yüksekliği")
plt.show
```

```python
##TÜM GÖRSELLERİN GENİŞLİĞİNİN DAĞILIM HİSTOGRAMI
plt.hist(img_width ,color='#0c457d' , bins = 10, histtype = "bar" , align = "mid")
plt.xlabel("Width")
plt.ylabel("Resim Adedi")
plt.title("Görsellerin Genişliğinin")
plt.show
```
